Plotting **Cluster Map** of our Data using **Hierarchircal Clustering**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Increase vertical length
plt.figure(figsize=(1, 300))  # Adjust the height as needed

# Assuming df_filled is your filled DataFrame
df_filled = result_df.fillna(0)

df_filled = df_filled[['IFNg_log2FCExp1', 'IFNg_log2FCExp2', 'IFNg_log2FCExp3',
                      'LNMA_log2FCExp1', 'LNMA_log2FCExp2', 'LNMA_log2FCExp3',
                      'IFNg_LNMA_log2fcEXP1', 'IFNg_LNMA_log2fcEXP2', 'IFNg_LNMA_log2fcEXP3']]
# Define custom colors for the colormap
colors = [(0, 0, 1), (1, 1, 1), (1, 0, 0)]  # Dark blue, white, dark red; adjusted to [0, 1] range

# Create a LinearSegmentedColormap
custom_cmap = LinearSegmentedColormap.from_list("custom_coolwarm", colors)

# Adjust vmin and vmax based on your data range
clustermap = sns.clustermap(df_filled, cmap=custom_cmap, vmin=-10, vmax=10, col_cluster=False, cbar_kws={"orientation": "horizontal", "pad": 10},
                            yticklabels=True, figsize=(15,28))
# Move the colorbar to the right
clustermap.cax.set_position([0.29, 0.84, 0.5, 0.02])  # Adjust the position and size as needed

plt.title("Genes Up/Downregulated on IFN-$\gamma$ Treatment", fontsize=15)
clustermap.ax_heatmap.set_ylabel('Genes', fontsize=15, fontweight='bold')
clustermap.ax_heatmap.set_xlabel('Treatment', fontsize=15, fontweight='bold', labelpad=15)
clustermap.cax.set_xlabel('Log$_2$ Fold Change', fontsize=14)
clustermap.ax_heatmap.set_xticklabels(['IFN-$\gamma$ (1)', 'IFN-$\gamma$ (2)', 'IFN-$\gamma$ (3)','LNMA (1)','LNMA (2)','LNMA (3)', 'IFN-$\gamma$ + LNMA (1)', 'IFN-$\gamma$ + LNMA (2)','IFN-$\gamma$ + LNMA (3)'],
                                      fontsize=15)  # Adjust the labels as needed
plt.xticks(fontsize=15)
# clustermap.ax_heatmap.set_xticklabels(clustermap.ax_heatmap.get_xticklabels(), fontsize=25)
# clustermap.ax_heatmap.xticks(fontsize=32)
plt.savefig("Genes_Up_or_Downregulated_on_IFNg_Treatment.png", dpi=500)
plt.show()

```
