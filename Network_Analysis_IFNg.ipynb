##Figure S27

Kernel Density Estimate Plots and Scatterplots investigating the relationship between network parameters

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats

# Extract the selected columns from joined_df
selected_columns = joined_df[['Degree', 'BetweennessCentrality', 'Mean_log2FC']]

# Calculate the Spearman correlation and p-values
spearman_corr, p_values = stats.spearmanr(selected_columns, axis=0)

# Create a figure and axes for the plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Define bin sizes for KDE plots
bin_sizes = [20, 30, 40]  # Adjust as needed

# Plot KDE distributions in the first row
for i, (col, bin_size) in enumerate(zip(['Degree', 'BetweennessCentrality', 'Mean_log2FC'], bin_sizes)):
    sns.histplot(selected_columns[col], kde=True, bins=bin_size, ax=axes[0, i], color='blue', linewidth=1, alpha=0.2)
    axes[0, i].set_xlabel('Degree', fontsize=16, fontweight='bold') if col == 'Degree' else axes[0, i].set_xlabel('Betweenness Centrality', fontsize=14, fontweight='bold') if col == 'BetweennessCentrality' else axes[0, i].set_xlabel('Log2 (Fold Change)', fontsize=16)
    axes[0, i].set_ylabel('Frequency', fontsize=16, fontweight='bold')  # Set y-axis label to "Frequency"
    axes[0, i].tick_params(axis='x', labelsize=16)  # Increase x tick label size
    axes[0, i].tick_params(axis='y', labelsize=16)  # Increase y tick label size
    axes[0, i].grid(False)  # Turn off grid

# Plot correlation plots in the second row
for i, (x_col, y_col) in enumerate([('BetweennessCentrality', 'Degree'), ('Mean_log2FC', 'BetweennessCentrality'), ('Degree', 'Mean_log2FC')]):
    sns.regplot(x=x_col, y=y_col, data=selected_columns, ax=axes[1, i], line_kws={"color":"red"})
    axes[1, i].set_xlabel('Betweenness Centrality', fontsize=16, fontweight='bold') if x_col == 'BetweennessCentrality' else axes[1, i].set_xlabel('Log2 (Fold Change)', fontsize=16, fontweight='bold') if x_col == 'Mean_log2FC' else axes[1, i].set_xlabel('Degree', fontsize=16, fontweight='bold')
    axes[1, i].set_ylabel('Degree', fontsize=16, fontweight='bold') if y_col == 'Degree' else axes[1, i].set_ylabel('\nBetweenness Centrality', fontsize=16, fontweight='bold') if y_col == 'BetweennessCentrality' else axes[1, i].set_ylabel('Log2 (Fold Change)', fontsize=16, fontweight='bold')
    p_val = p_values[i, (i + 1) % 3]
    if p_val < 10e-4:
        p_val_str = 'p < 10â»â´'
    else:
        p_val_str = f'p = {p_val:.4f}'
    axes[1, i].text(0.7, 0.9, f"Spearman: {spearman_corr[i, (i + 1) % 3]:.2f}\n{p_val_str}",
                    ha='center', va='center', transform=axes[1, i].transAxes, fontsize=16, color='black')
    axes[1, i].tick_params(axis='x', labelsize=16)  # Increase x tick label size
    axes[1, i].tick_params(axis='y', labelsize=16)  # Increase y tick label size
    axes[1, i].grid(False)  # Turn off grid

# Adjust layout
plt.tight_layout()
plt.savefig("Network_paramsters_scatter_.png", dpi=500)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Filter the data for hubs_only and select the required columns
hubs_only_data = hubs_only[['Name (Yellow hubs)', 'Mean_log2FC', 'BetweennessCentrality', 'Degree']]

# Sort the data by 'Degree' in ascending order
hubs_only_data = hubs_only_data.sort_values(by='Degree', ascending=False)

# Set the style
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=hubs_only_data, x='Name (Yellow hubs)', y='Mean_log2FC', hue='Degree', palette='coolwarm',s =100)  # Adjust 's' as needed

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right',fontsize=16)
plt.yticks(fontsize=16)

# Set labels and title
plt.xlabel('Hub Names', fontsize=16,fontweight='bold')
plt.ylabel('Log2 (Fold Change)', fontsize=16, fontweight='bold')
# plt.title('Scatter Plot of Mean_log2FC vs Name (Yellow hubs)', fontsize=14)

# Show legend
plt.legend(title='Degree', fontsize=14)

# Calculate Spearman correlation and p-value
spearman_corr, p_value = stats.spearmanr(hubs_only_data['Mean_log2FC'], hubs_only_data['BetweennessCentrality'])

# Annotate the plot with Spearman correlation coefficient and p-value
plt.text(0.2, 0.8, f"Spearman: {spearman_corr:.2f}\n p-value: {p_value:.4f}",
         ha='center', va='center', transform=plt.gca().transAxes, fontsize=16, color='black', bbox=dict(facecolor='white', alpha=0.50))


# 'Degree', fontsize=16, fontweight='bold'
# Show plot
plt.tight_layout()
plt.savefig("Log2_Fold_change_vs_Hub_Names.png", dpi=500)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
import powerlaw

# Extract the 'Degree' column from joined_df
degree_data = joined_df['Degree']
sns.axes_style("ticks")
# Calculate KDE for Degree
sns.histplot(degree_data, kde=True, color='blue', linewidth=1, alpha=0.2)
plt.xlabel('Degree', fontsize=16, fontweight='bold')
plt.ylabel('Frequency', fontsize=16, fontweight='bold')
plt.tick_params(axis='both', labelsize=16)

# Display statistics
plt.title('Power Law of Degree Distribution', fontsize=16)

# Fit a power-law distribution to the data
fit = powerlaw.Fit(degree_data)

# Get the exponent of the power-law fit
exponent = fit.alpha

# Get R^2 and p-value
R2, p_value = fit.distribution_compare('power_law', 'exponential', normalized_ratio=True)

# Display KS Score, exponent of power law fit, R^2, and p-value
plt.text(0.7, 0.8, f"\nKS Score: {fit.D:.2f}\nExponent: {exponent:.2f}\nR$^2$: {R2:.2f}\np-value: {p_value:.4f}",
         ha='left', va='center', transform=plt.gca().transAxes, fontsize=12, color='black')



plt.tight_layout()
plt.savefig("Power_Law_of_Degree_Distribution.png", dpi=500)
# plt.grid(False)
plt.show()
```

```python
from igraph import Graph
import numpy as np
import pandas as pd

# Set up an empty list to store details for each network
network_details = []

# Use a for loop to repeat 1000 times
for i in range(1000):
    # Generate a random network using the Erdos-Renyi model
    g = Graph.Erdos_Renyi(n=78, m=275, directed=False)

    # Calculate clustering coefficient of this network
    cc = g.transitivity_undirected()

    # Calculate the average shortest path length of this network
    aspl = g.average_path_length()

    # Store details for this network in a dictionary
    network_details.append({
        "Network": i + 1,
        "Clustering Coefficient": cc,
        "Average Shortest Path Length": aspl
    })

    # Print results
    print("Network:", i + 1)
    print("Clustering Coefficient:", cc)
    print("Average Shortest Path Length:", aspl)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(network_details)

# Calculate mean values
cc_mean = df["Clustering Coefficient"].mean()
aspl_mean = df["Average Shortest Path Length"].mean()

print("Mean Clustering Coefficient:", cc_mean)
print("Mean Average Shortest Path Length:", aspl_mean)

# Write DataFrame to a CSV file
# df.to_csv("random_network_details.csv", index=False)

df
```
